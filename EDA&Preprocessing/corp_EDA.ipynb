```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_rows = 500    
pd.options.display.max_columns = 20
```

```python
corp = pd.read_csv('company1.csv')
```

```python
corp.drop(['Unnamed: 0','Unnamed: 0.1'], axis=1, inplace=True)
corp
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
plt.rcParams["axes.unicode_minus"] = False
rc("font", family= "Malgun Gothic")
%matplotlib inline
```

# EDA

### 기업구분에 따른 회사 수

```python
sns.countplot(x='기업구분2', data=corp);
```

- 중소기업의 수가 압도적으로 많다. 

### 기업구분에 따른 사원수

```python
plt.figure(figsize=(16,6))
sns.boxplot(x='사원수2', y='기업구분2', data=corp)
plt.show()
```

- 중견기업, 대기업, 기타 ==> 이상치 제거 필요
- 중소기업은 max로 해서 가장 큰 값 두개 만 제거하자

```python
# 별점의 분포가 어떤지 : countplot 은 데이터의 개수를 표시하는 플롯
plt.figure(figsize=(12, 8))
sns.countplot(data=corp, x='별점');
```

### 기업구분에 따른 별점 상자그래프

```python
plt.figure(figsize=(6,4))
# sns.boxplot(x='별점', y='기업구분2', data=corp)
sns.boxplot(x='기업구분2', y='별점', data=corp)
plt.show()
```

- 별점은 이상치 제거하면 안될것 같다.
- 별점은 이상치가 더 중요한 데이터 값 --> 과연?

```python
# 별점별 사원수 
plt.figure(figsize=(12, 8))
sns.barplot(x='별점', y='사원수2', data=corp);
```

```python
# 입사율,퇴사율 수치만 가져와서 확인
corp['입사율(%)'] = corp['입사율'].str.split().str[0]
corp['퇴사율(%)'] = corp['퇴사율'].str.split().str[0]
corp['입사율(%)'] = corp['입사율(%)'].astype('float')
corp['퇴사율(%)'] = corp['퇴사율(%)'].astype('float')
corp.info()
```

### 입사율 및 퇴사율과 별점 사이의 관계 확인

```python
sns.pairplot(corp[['별점','입사율(%)','퇴사율(%)']])
```

###  지역(도) 기준으로 별점 분포 확인 

```python
corp['지역(도)'] = corp['지역'].str.split().str[0]
sns.barplot(data=corp, x= '지역(도)', y="별점")
```

### 수치형 자료로 확인한 상관관계 히트맵

```python
plt.figure(figsize=(12,6))
sns.heatmap(corp.corr(), annot=True, fmt = '.2f', lw=0.2, cmap = 'YlGnBu')
plt.show()
# plt.title('',fontsize= 20);
```
